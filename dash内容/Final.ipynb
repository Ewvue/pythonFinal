# 总结

近年来我国各省的结婚人数普遍呈下降的趋势，我们主要是通过经济方面（各省人均GDP，各省商品房房价，单元员工平均工资）以及受教育程度（本科毕业人数）来分析近十年来结婚人数减少的原因，剖析原因及其危害，并为其提出建议。

从以上的分析中我们可以看到，中国经济越发达地区的结婚率越低，这主要是因为经济发展，导致物价，房价等人们生活的成本提高，人口生活压力增大。同时结婚的成本增大，如一些“没有稳定的工作和收入不结婚”“没房没车不结婚”等都成了结婚的附加条件，与此同时，经济发展，经济发达的地区的房价不断地上升，导致购房成为了年轻人的一大难题。许多年轻人也因此成为“房奴”等。这些经济因素都会导致年轻人对结婚这个问题望而止步。同时，随着经济的发展，人们受教育的程度不断提高，人们的婚姻观念的改变，不仅是以“成家”为主，更多的倾向于“立业”，不断地充实自己，追求事业上的成功，不愿意早早地结婚束缚住自己的发展。
结婚人数不断地减少的社会经济发展进步的一个必然结果，但是结婚屡创新低，对社会发展也存在不利影响。一个国家的繁荣与发展，首先有赖于人口的稳定和发展，结婚率走低，带来的直接影响就是出生率的降低，以及人口代际的延长。比如，25岁结婚，26岁生育，百年之内基本上将产生四代人。但是35岁结婚，36岁生育，百年之内将只有三代人，时间越长，人口出生率越低，这就是现实。 而人口老龄化的问题也随之涌现，老年人口的抚养问题等对年轻人也是一种压力。

针对以上问题的建议是虽然经济问题是一个很大的压力，但是也不要因为经济的原因而耽误了自己的婚姻幸福，应树立正确的社会导向，把成家立业、生儿育女、快乐生活当作人生乐趣，形成婚姻家庭的良性循环。
